**Work with Color-Images**
**and how to transform Color-Images to Grayscale**

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = plt.imread("images/fibers2.png")
plt.figure()
plt.imshow(img)
plt.show()
```

This image is composed of three color channels, and one alpha channel. (Sometimes there is no alpha channel) Now if we want to know how many channels there are, we can use the numpy shape command.

We can extract the Colors with the same numpy indexing. Red is the first channel (Index 0), Blue is the second channel (Index 1), and Green is the third channel (Index 2).

```python
plt.figure(figsize = (12,8))
plt.subplot(131)
plt.title("Red")
plt.imshow(img[:,:,0],cmap = "Reds")

plt.subplot(132)
plt.title("Green")
plt.imshow(img[:,:,1],cmap = "Greens")

plt.subplot(133)
plt.title("Blue")
plt.imshow(img[:,:,2],cmap = "Blues")

plt.tight_layout()
plt.show()
```

If we want to know only the amount of light intensity in every pixel (without the color), we can just add all the channels together. The problem with this approach is that the human eye is much more sensitive to green light, so we loose contrast. To avoid this we can use a weighted sum of the channels (CIE 1931).

```python
grayscale_image = img[:,:,0] + img[:,:,1] + img[:,:,2]
grayscale_image_adapted_to_human_vision = 0.21 * img[:,:,0] + 0.72 * img[:,:,1] + 0.07 * img[:,:,2]

plt.figure(figsize = (12,8))
plt.subplot(121)
plt.title("Sum")
plt.imshow(grayscale_image,cmap = "gray")
plt.subplot(122)
plt.title("Weighted Sum")
plt.imshow(grayscale_image_adapted_to_human_vision,cmap = "gray")
plt.show()
```
