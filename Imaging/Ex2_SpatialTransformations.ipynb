Spatial Transformations : Image Negatives, Log Transformation, Power-Law Transformation, Contrast Stretching

```python
import numpy as np
import matplotlib.pyplot as plt
```

1. Linear function : Image negatives

```python
img = plt.imread('images/blumen.jpg')
normImg=np.divide(np.double(img),255)
img_negative = 255 - img
plt.figure(figsize = (10,10))
plt.subplot(221)
plt.imshow(img,cmap='gray')
plt.title('Original')
plt.subplot(222)
plt.imshow(img_negative,cmap='gray')
plt.title('Negative Image')
plt.subplot(223)
plt.hist(img.ravel(),256,[0,256])
plt.subplot(224)
plt.hist(img_negative.ravel(),256,[0,256])
plt.show()
```

2. Log Transformation

```python
c = 255 / (np.log(1 + np.max(normImg))+ 0.0000001)
logImage = c * (np.log(normImg + 1))
# Display both images
plt.figure(figsize = (10,10))
plt.subplot(221)
plt.imshow(img)
plt.title('Original')
plt.subplot(222)
plt.imshow(logImage.astype(np.uint8))
plt.title('Log Transformed Image')
plt.subplot(223)
plt.hist(img.ravel(),256,[0,256])
plt.subplot(224)
plt.hist(logImage.ravel(),256,[0,256])
plt.show()
```

3.Power-law (gamma) transformations 
Image is dark when gamma > 1 and Bight when gamma < 1 

```python
imgE = plt.imread('images/AlbertEinstein.jpg')
normImgE=np.divide(np.double(imgE),255)
c,r=1,5
imgGamma1 = c * np.power(normImgE, r) # It makes the image dark
r1=0.4
imgGamma2 = c * np.power(normImgE, r1) #It makes the image bright
plt.figure(figsize = (15,15))
plt.subplot(131)
plt.imshow(imgE,cmap='gray')
plt.title('Original')
plt.subplot(132)
plt.imshow(imgGamma1,cmap='gray')
plt.title('PowerLaw Transformed Image')
plt.subplot(133)
plt.imshow(imgGamma2,cmap='gray')
plt.title('PowerLaw Transformed Image')
plt.show()
```

4. Contrast Stretching

```python
constant = (255-0)/(normImgE.max()-normImgE.min())
img_stretch = normImgE * constant
plt.figure(figsize = (10,10))
plt.subplot(221)
plt.imshow(normImgE)
plt.title('Original')
plt.subplot(222)
plt.imshow(img_stretch.astype(np.uint8))
plt.title('Contrast Strecthed Image')
plt.subplot(223)
plt.hist(normImgE.ravel(),256,[0,256])
plt.subplot(224)
plt.hist(img_stretch.ravel(),256,[0,256])
plt.show()
```
