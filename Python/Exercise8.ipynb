```python
from sympy import *
```

```python
x = Symbol('x')
f = 2*x + 8 + 66/(x-8)
print('Simplified f(X) = %s :' %f)
f.simplify()
```

```python
plot(f,(x,0,4),title='f(x) in [0;4]',show=True)
```

```python
plot(f,(x,0,20),title='f(x) in [0;20]',show=True)
```

```python
polyRoots = solveset(f, x)
print('Zeros of f(x) = %s : %s'%(f,polyRoots))
print('Zeros of f(x) =%s in decimal form is %s' %(f,polyRoots.evalf()))
```

```python
fprime = f.diff(x)
print('Derivative of f(x) : ' ,simplify(fprime))
```

```python
d2f_dxx = f.diff(x,2)
print('Second Derivative f''(X): ',d2f_dxx.simplify())
d2Roots = solve(d2f_dxx)
dRoots = solveset(fprime, x)
print('Critical Points fprime (in decimal form) is ', dRoots.evalf())
```

```python
interval = [[0, 4], [0, 20]]
for j in interval:
    print('\nExtrema of the function f(x) = %s defined in interval %s : ' %(f,j))
    #Check whether the critical points lie inside the given interval [0;4]
    criticalPts = list(filter(lambda x: x >= j[0] and x <= j[1], dRoots))
    print('\tThe Critical Points %s lie inside the interval %s' %(criticalPts,j))
    #Finding the extrema
    for i in dRoots:
        if i not in criticalPts:
            print('\t',i, 'is not local extrema')
        else:
            if d2f_dxx.subs(x,i)>0:
               print('\tLocal Minima : %s at x = %s' %(f.subs(x,i).evalf(),i))
            elif d2f_dxx.subs(x,i)<0:
                 print('\tLocal Maxima : %s at x = %s' %(f.subs(x,i).evalf(),i))
            else:
                 print('\t',criticalPts[i], 'is not local extrema')
```

```python
I = integrate(f,(x,polyRoots.args[0],polyRoots.args[1]))
print('Integral of f(X) between both zeros %s : %f' %(simplify(polyRoots),I.evalf()))
I.simplify()
```
